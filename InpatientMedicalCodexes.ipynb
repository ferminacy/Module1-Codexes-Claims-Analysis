```python
#Loading the data
df = pd.read_csv('https://data.cms.gov/sites/default/files/2023-04/67157de9-d962-4af0-bf0e-3578b3afec58/inpatient.csv',sep= '|' )
```

```python
#Exploring dataset
df.sample(10)
```

```python
df.columns
```

```python
#Explore the dataset to identify columns related to medical codexes
codex_columns = [col for col in df.columns if 'ICD' in col or 'HCPCS' in col or 'DRG' in col]
print("Medical Codex Columns:", codex_columns)
```

```python
#Calculating frequency
for col in codex_columns:
    print(f"\nFrequency of values in {col}:")
    print(df[col].value_counts())
```

```python
#Determine if there are any missing or null values in the codex-related columns
df[codex_columns].fillna('Unknown', inplace=True)
print(df[codex_columns].isnull().sum())
```

```python
#Common codes in the dataset
for col in codex_columns:

    value_counts = df[col].value_counts()

    top_codes = value_counts.head(10).index.tolist()

    print(f"\nTop codes in {col}:")
    print(top_codes)
```
